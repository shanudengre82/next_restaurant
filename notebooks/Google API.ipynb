```python
lst_ = []


for item in result:
    name_ = item['name']
    lst_.append(name_.split())

#for restaurants_names in :
      #  lst_.append(restaurants_names)
lst_
```

```python
# remove 'restaurant, bar, lounge' from restaurant name

to_remove = ['Restaurant', 'Bar', 'Lounge']

for items in lst_:
    for word in items:
        for toremove in to_remove:
            if word == toremove:
                items.remove(word)

clean_lst = []

for items in lst_:
    if '-' in items:
        index = items.index('-') 
        clean_lst.append(items[:index])
    else:
        clean_lst.append(items)
        
# cut out anything that comes after a '-' in a restaurant name
        
clean_name = ''
name_lst = []

for item in clean_lst:
    for word in item:
        clean_name += word +' '
    name_lst.append(clean_name.rstrip())
    clean_name = ''
    
```

```python
# add cleaned restaurant names to the df

df['name'] = name_lst
        
```
