## API REQUEST


```python
import requests
import pandas as pd
```

```python
def fetch_here(lat, lon, api_key):
    
    url = f'https://discover.search.hereapi.com/v1/discover?at={lat},{lon}&q=restaurant&limit=100&apiKey={api_key}'
    response = requests.get(url)
    
    if response.status_code != 200:
        return ''
    data = response.json()
    
    return data['items']
```
